```python
def factorial(num):
    if (num <= 0):
        return 1

    return num * factorial(num - 1)
```

```python
def tailor(val, power):
    iter = 0

    for k in range(0, power):
        iter += ((-1) ** k) * ((val ** (2*k+1)) / factorial(2*k+1))
        #print(iter)

    #print('')
    return iter
```

```python
for power in [1,3,7]: # По условию
    plot_x = np.arange(0, np.pi*2, 0.1)
    plot_y = list(map(lambda a: tailor(a, power), plot_x))

    plt.title(f'Крайняя степень ряда = {power}')
    plt.plot(plot_x, plot_y)
    plt.show()
```
