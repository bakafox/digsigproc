```python
import numpy as np
from scipy.integrate import quad, tplquad
import matplotlib.pyplot as plt
```

### Стр. 58, 8.1, задание 1

```python
def fun(x, y, z):
    return (x + y + z) / np.sqrt(2 * x ** 2 + 4 * y ** 2 + 5 * z ** 2)

def border_1(x):
    return np.sqrt(1 - x ** 2)

def border_2(x, y):
    return np.sqrt(1 - x ** 2 - y ** 2)
```

```python
[res_1, err_1] = tplquad(fun, 1, 0, border_1, 0, border_2, 0)
print(f'Результат: {res_1}\nОшибка выч.: {err_1}')
```

### Стр. 59, 8.1, задание 3

```python
def impulse(x):
    A = 2

    if (x < 0 or x > 1):
        return 0

    return A
```

```python
plot_x = np.arange(-1, 2, 0.01)
plot_y = list(map(lambda a: impulse(a), plot_x))
plt.plot(plot_x, plot_y)
plt.show()
```

```python
[res_3, err_3] = quad(impulse, -100, 100)
print(f'Результат: {res_3}\nОшибка выч.: {err_3}')
```
